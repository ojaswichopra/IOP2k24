```python
y = df['Severity_None']
y
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
print('X_train: ', len(X_train))
print('X_test:', len(X_test))
```

```python
from collections import Counter
Counter(y_train)
```

```python
Counter(y_test)
```

### KNN Model

#### Building the Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

#### Training the Model

```python
knn.fit(X_train,y_train)
```

#### Finding Performance Metrics

```python
ypred=knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report

ac=accuracy_score(y_test,ypred)
cm=confusion_matrix(y_test,ypred)
cr=classification_report(y_test,ypred)

print(f"{ac}\n\n{cm}\n\n{cr}")
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.title('Confusion Matrix')
plt.savefig('asthma_cm.png')
plt.show()
```

#### Saving a model

```python
import pickle
filename = 'asthma.sav'
pickle.dump(knn, open(filename, 'wb'))
```

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors

# ... your data and KNN model setup

# Create a dense grid
x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x1_min, x1_max, 0.1),
                     np.arange(x2_min, x2_max, 0.1))

# Predict classes for the grid points
knn = NearestNeighbors(n_neighbors=5)  # Your KNN model 
knn.fit(X)
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm) 
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KNN Decision Boundaries')
plt.show()
```
